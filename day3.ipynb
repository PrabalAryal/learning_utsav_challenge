```python
cap_data=data.copy()
```

```python
cap_data
```

```python
def cappingoutliers(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    iqr=q3-q1
    upper_limit=q3+(1.5*iqr)
    lower_limit=q1-(1.5*iqr)
    data.loc[data>upper_limit]=upper_limit
    data.loc[data<lower_limit]=lower_limit
    return data
cols_to_cap=["ALB","ALP","ALT","AST","BIL","CHE","CHOL","CREA","GGT","PROT"]

for i in cols_to_cap:
    cap_data[i]=cappingoutliers(cap_data[i])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cap_data['Category']=LabelEncoder().fit_transform(cap_data['Category'])
```

```python
cap_data
```

```python
cap_data['Sex']=cap_data['Sex'].apply(lambda x: 1 if x=='m' else 0)
```

```python
cap_data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
cols_to_scale=["Age","ALB","ALP","ALT","AST","BIL","CHE","CHOL","CREA","GGT","PROT"]
cap_data[cols_to_scale]=scaler.fit_transform(cap_data[cols_to_scale])
```

```python
cap_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x=cap_data.drop(['Category'],axis=1)
y=cap_data['Category']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=SVC()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
accuracy_score(y_test,prediction)
```

```python
print(classification_report(y_test,prediction))
```

```python
confusion_matrix(y_test,prediction)
```
