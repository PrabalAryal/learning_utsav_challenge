```python
#outliers removal
```

```python
def outliercapping(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    iqr=q3-q1
    upperlimit=q3+1.5*iqr
    lowerlimit=q1-1.5*iqr
    data.loc[data>upperlimit]=upperlimit
    data.loc[data<lowerlimit]=lowerlimit
    return data

cols=['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)',
       'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH',
       'AH']
for i in cols:
    data[i]=outliercapping(data[i])
```

```python
#checking for remaining outliers
rem_cols=['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)',
       'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH',
       'AH']
plt.figure(figsize=(10,10))
sns.boxplot(data=data[rem_cols],orient='v',palette='Set2')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
cols=['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)',
    'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH',
    'AH']
data[cols] = StandardScaler().fit_transform(data[cols])

```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,accuracy_score,confusion_matrix,classification_report
```

```python
x=data.drop(['T','RH','AH'],axis=1)
y=data[['T','RH','AH']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model_T = SVR()
model_T.fit(x_train, y_train['T'])

model_RH = SVR()
model_RH.fit(x_train, y_train['RH'])

model_AH = SVR()
model_AH.fit(x_train, y_train['AH'])
```

```python

predictions_T = model_T.predict(x_test)
predictions_RH = model_RH.predict(x_test)
predictions_AH = model_AH.predict(x_test)
print('predictions_T:',predictions_T)
print('predictions_RH:',predictions_RH)
print('predictions_AH:',predictions_AH)
```

```python
print(mean_squared_error(y_test['T'], predictions_T))
print(mean_squared_error(y_test['RH'], predictions_RH))
print(mean_squared_error(y_test['AH'], predictions_AH))
```
